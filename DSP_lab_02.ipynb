```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
```

APPLICATIONS OF CONVULATION

```python
# Create signal ...
Srate = 256 #Hz Sample rate
t = np.arange(0,3,1/Srate)

pnts = len(t)
```

```python
# Creating a noiseless signal
x = np.sin(2*np.pi*2*t)
#Creating a random noise
noise = 5*np.random.randn(pnts)

#Adding noise to original signal
NoisySignal = x + noise

```

```python
#Plotting original signal

plt.figure(figsize=(20,12))
style.use('dark_background')

plt.plot(t,x, label='Original Signal')
plt.legend(fontsize=30)

plt.show()
```

```python
#Plotting noisy signal
plt.figure(figsize=(20,12))
style.use('dark_background')

plt.plot(t,NoisySignal, label='Noisy Signal')
plt.legend(fontsize=30)

plt.show()
```

```python
#Plotting noisy signal
plt.figure(figsize=(20,12))
style.use('dark_background')

plt.stem(t,NoisySignal, label='Noisy Signal Discrete')
plt.legend(fontsize=30)

plt.show()
```

```python
#selecting filters/kernel/impulse for convulation

filter = 2*np.ones(50)/10.0
print(filter)
```

```python
#Filtering/ Noise-Canllation by Convulation
filteredSignal = np.convolve(NoisySignal, filter, mode = 'full')
```

```python
#Plotting Filtered signal
plt.figure(figsize=(20,12))
style.use('dark_background')

plt.plot(filteredSignal, label='Filtered Signal')
plt.legend(fontsize=30)

plt.show()
```

```python
#double filter

filteredSignal2 = np.convolve(filteredSignal, filter, mode = 'full')
```

```python
#Plotting Filtered signal
plt.figure(figsize=(20,12))
style.use('dark_background')

plt.plot(filteredSignal2, label='Double Filtered Signal')
plt.legend(fontsize=30)

plt.show()
```

```python
filteredSignal3 = np.convolve(filteredSignal, filter, mode = 'full')

#Plotting Filtered signal
plt.figure(figsize=(20,12))
style.use('dark_background')

plt.stem(filteredSignal3, label='Tripple Filtered Signal')
plt.legend(fontsize=30)

plt.show()
```

Wizney Method

```python
#wizney method

x = np.array([1,2,3,1,3,4,6])
h = np.array([1,2,1,-1])

metric = np.zeros(len(x)*len(h)).reshape(len(x),len(h))


#Metric 
for i in range(len(x)):
  for j in range(len(h)):
    metric[i,j]=x[i]*h[j]

print(metric)
```

```python
signal = np.zeros((len(x)+len(h))-1)

for i in range((len(x)+len(h))-1):
    for j in range(len(x)):
        for k in range(len(h)):
            if(j+k == i):
                signal[i] += metric[j][k]
                
print(signal)  
```
