```python
#Drawing unit sample sequence

import matplotlib.pyplot as plt
import numpy as np

n = np.arange(-10,11)

#unit sample sequence
unit_sample = np.zeros_like(n)
unit_sample[n==0] = 1

#unit Step sequence
unit_step = np.heaviside(n,0)

#unit ramp sequence
unit_ramp = np.maximum(n,1)

#ploting the signals

fig, axs = plt.subplots(3,1,figsize=(8,8))

axs[0].stem(n, unit_sample)
axs[0].set_title('Unit Sample Sequence')

axs[1].stem(n, unit_step)
axs[1].set_title('Unit Step Sequence')

axs[2].stem(n, unit_ramp)
axs[2].set_title('Unit Ramp Sequence')

plt.tight_layout()
plt.show()
```
