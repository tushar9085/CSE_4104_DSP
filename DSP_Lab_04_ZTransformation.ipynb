```python
# Z transformation

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
def plot_pole_zero(poles, zeros):
  fig, ax = plt.subplots()

  #Ploting poles
  ax.plot(np.real(poles), np.imag(poles), 'rx', markersize = 10, label = "Poles")
  #Ploting Zeros
  ax.plot(np.real(zeros), np.imag(zeros), 'rx', markersize = 10, label = "Zeros")
  
  #Set plotting attributes
  ax.axhline(0, color="black", linewidth=0.5)
  ax.axvline(0, color="black", linewidth=0.5)
  ax.grid(True, which = 'both', linestyle = '--',linewidth = 0.5)
  ax.legend()

  ax.set_xlabel('Real')
  ax.set_ylabel('Imaginary')

  ax.set_title("Pole-Zero Diagram")

  #Plotting limits
  ax.set_xlim([-2,2])
  ax.set_ylim([-2,2])

  plt.show()



poles = [-0.5+0.5j, -0.5-0.5j, -1.0]
zeros = [0.2j, -0.2j]  #Complex conjugate

plot_pole_zero(poles, zeros)
```

```python

def Z_Transformation(signal, n):
  N = len(signal)
  Z = np.zeros(N, dtype=complex)   #Array make korbe, shob zero thakbe but complex zero

  for k in range(N):
    Z[k] = np.sum(signal * np.exp(-1j*2*np.pi*n*k/N))

  roc = 'Entire Z-Plane'
  if np.abs(Z).any():
    roc = 'ROC: |Z| < {}'.format(np.max(np.abs(Z)))

  return Z, roc


signal = [1,2,3,4,5]
n = np.arange(len(signal))

Z, roc = Z_Transformation(signal, n)


#Ploting ROC

fig, ax = plt.subplots()
circle = plt.Circle((0,0), radius = np.max(np.abs(Z)), edgecolor = 'yellow')
ax.add_patch(circle)
plt.xlim(-np.max(np.abs(Z))-1, np.max(np.abs(Z))+1)
plt.ylim(-np.max(np.abs(Z))-1, np.max(np.abs(Z))+1)


plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.title("Region of COnvergence")
plt.grid(True, linestyle = '--', linewidth = 0.5)
plt.show()



#printing the values of Z
print('Z- Transformation Values---')
for i, z in enumerate(Z):
  print('Z[{}]: {}'.format(i, z))

print(roc)
```
