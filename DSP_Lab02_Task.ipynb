```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
import math
```

```python
fs = 100
n_samples = 10

t = np.arange(0,n_samples) 

# Generate the sine wave signal
signal1 = np.sin(2 * np.pi * t)
```

```python
print(signal1)
```

```python
# Plot the signal1
plt.plot(t, signal1)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Signal1')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fs = 500
n_samples = 10

t = np.arange(n_samples) 

# Generate the sine wave signal
signal2 = np.sin(2  * t)
```

```python
print(signal2)
```

```python
# Plot the signal
plt.plot(t, signal2)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Signal2')
plt.show()
```

```python
x = signal1
h = signal2

#x = np.array([1,2,3,1])
#h = np.array([1,2,1,-1])

metric = np.zeros(len(x)*len(h)).reshape(len(x),len(h))

#Metric 
for i in range(len(x)):
  for j in range(len(h)):
    metric[i,j]=x[i]*h[j]

print(metric)
```

```python
signal = np.zeros((len(x)+len(h))-1)
signal_len = (len(x)+len(h))-1
#print(signal_len)
range_find = math.floor(signal_len/2)


for i in range(-range_find, range_find+1):
    for j in range(len(x)):
        for k in range(len(h)):
            if(k-j == i):
                signal[i+range_find] += metric[j,k]
                         
print(signal) 
```
