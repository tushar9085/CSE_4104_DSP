```python
import numpy as np
import matplotlib.pyplot as plt

fs = 1000
n_samples = 50

t = np.arange(n_samples) 

# Generate the sine wave signal
signal1 = np.sin(2 * np.pi * t)
```

```python
print(signal1)
```

```python
# Plot the signal1
plt.plot(t, signal1)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Signal1')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fs = 1000
n_samples = 50

t = np.arange(n_samples) 

# Generate the sine wave signal
signal2 = np.sin(2  * t)
```

```python
print(signal2)
```

```python
# Plot the signal
plt.plot(t, signal2)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Signal2')
plt.show()
```

```python
x = signal1
h = signal2

metric = []


#Metric 
for i in range(len(h)):
  row = []
  for j in range(len(x)):
    row.append(h[i]*x[j])
  
  metric.append(row)

print(metric)
```
